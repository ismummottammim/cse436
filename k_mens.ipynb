**Title**

K-Means Clustering with Elbow Method for Optimal Cluster Selection

**Objective**

To group unlabeled data points into clusters using K-Means algorithm
To determine the optimal number of clusters (K) using the Elbow Method
To visualize clustering results and cluster centroids

kmeans

```python
import matplotlib.pyplot as plt

x = [0,0,5,5,4,1,4,0]
y = [0,1,4,5,5,0,2,4]

plt.scatter(x, y)
plt.show()
```

```python
from sklearn.cluster import KMeans

data = list(zip(x, y))
inertias = []

for i in range(1, len(data) + 1):
    kmeans = KMeans(n_clusters=i, random_state=0, n_init='auto')
    kmeans.fit(data)
    inertias.append(kmeans.inertia_)

plt.plot(range(1, len(data) + 1), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(data)

plt.scatter(x, y, c=kmeans.labels_)
plt.show()
```

cluster

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans


x = [0,0,5,5,4,1,4,0]
y = [0,1,4,5,5,0,2,4]

data = list(zip(x, y))


plt.scatter(x, y)
plt.title("Original Data Points")
plt.show()


inertias = []

for i in range(1, 9):
    kmeans = KMeans(n_clusters=i, random_state=0, n_init='auto')
    kmeans.fit(data)
    inertias.append(kmeans.inertia_)

plt.plot(range(1, 9), inertias, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()


kmeans = KMeans(n_clusters=2, random_state=0, n_init='auto')
kmeans.fit(data)

plt.scatter(x, y, c=kmeans.labels_, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:,0],
            kmeans.cluster_centers_[:,1],
            color='red', marker='X', s=200)

plt.title("K-Means Clustering (2 Clusters)")
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

x = [0,0,5,5,4,1,4,0]
y = [0,1,4,5,5,0,2,4]

data = list(zip(x, y))


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))


ax1.scatter(x, y)
ax1.set_title("Original Data Points")

inertias = []
for i in range(1, 9):
    kmeans = KMeans(n_clusters=i, random_state=0, n_init='auto')
    kmeans.fit(data)
    inertias.append(kmeans.inertia_)

ax2.plot(range(1, 9), inertias, marker='o')
ax2.set_title('Elbow Method')
ax2.set_xlabel('Number of Clusters')
ax2.set_ylabel('Inertia')

plt.tight_layout()
plt.show()


kmeans = KMeans(n_clusters=2, random_state=0, n_init='auto')
kmeans.fit(data)


plt.figure(figsize=(8, 6))
plt.scatter(x, y, c=kmeans.labels_, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:,0],
            kmeans.cluster_centers_[:,1],
            color='red', marker='X', s=200)
plt.title("K-Means Clustering (2 Clusters)")
plt.show()
```

**Discussion**

K-Means works well for small, well-separated datasets like this
The algorithm is sensitive to K selection, hence Elbow Method is important
Results depend on:
Initial centroid placement
Value of K
Limitations:
Assumes clusters are spherical
Not ideal for irregular shapes or overlapping clusters
Visualization confirms that clustering is meaningful and interpretable
